```python
#load iris dataset
import seaborn as sns
data=sns.load_dataset('iris')
data
```

```python
#Drop the species column since this is a clustering problem.
data=data.drop('species',axis=1)
data
```

```python
#Scaling features
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
data_scaled=scalar.fit_transform(data)
data_scaled
```

Clustering Algorithm Implementation


1) KMeans Clustering

```python
from sklearn.cluster import KMeans  # Import the KMeans class
# Elbow method to find optimal number of clusters
Wcss=[]
k_values=range(1,11)
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(data_scaled)
  Wcss.append(kmeans.inertia_)
  #plotting elbow curve
import matplotlib.pyplot as plt
plt.plot(k_values,Wcss,marker="*")
plt.xlabel("Number of Clusters")
plt.ylabel("Wcss")
plt.title("Elbow  Curve")
plt.show()
```

The elbow method helps us visually estimate the optimal number of clusters (k) by identifying the point on the curve where the decrease in WCSS (Within-Cluster Sum of Squares) begins to slow down, resembling an elbow.

From this generated elbow curve,the "elbow" appears to be around k=3. This suggests that 3 clusters might be a good choice for your dataset.

```python
#k-mean
kmeans=KMeans(n_clusters=3,init='k-means++',n_init=15)
data['Cluster']=kmeans.fit_predict(data_scaled)
data
```

```python
# Visualize the clusters using seaborn
sns.pairplot(data, hue='Cluster', palette='viridis')
plt.show()
```

```python
#Silhoutte score
from sklearn.metrics import silhouette_score
silhouette_avg=silhouette_score(data_scaled,data['Cluster'])
print("Silhouette Score : ",silhouette_avg)
```

```python
#Creating profile
cluster_profile=data.groupby("Cluster").mean()
cluster_profile
```

KMeans clustering is an unsupervised learning algorithm that aims to partition data points into k distinct clusters based on their similarity. Here's a breakdown of the process:

1) Initialization: The algorithm randomly selects k points in the dataset as initial cluster centroids.

2)Assignment: Each data point is assigned to the nearest centroid based on a distance metric (typically Euclidean distance).

3)Update: The centroids of each cluster are recalculated as the mean of all data points assigned to that cluster.

4) Iteration: Steps 2 and 3 are repeated iteratively until the centroids stabilize or a maximum number of iterations is reached. The result is k clusters, where each data point belongs to the cluster with the closest centroid.

#suitability for iris
The Iris dataset's characteristics—distinct groups, numerical features, and relatively compact clusters—align well with the assumptions and workings of the KMeans algorithm, making it a suitable clustering technique for this dataset.





2)Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3,linkage='ward')
data['Cluster']=hc.fit_predict(data_scaled)
data
```

```python
# Visualize the clusters using seaborn
sns.pairplot(data, hue='Cluster', palette='viridis')
plt.show()
```

```python
#Silhoutte score
from sklearn.metrics import silhouette_score
silhouette_avg=silhouette_score(data_scaled,data['Cluster'])
print("Silhouette Score : ",silhouette_avg)
```

```python
#Creating profile
cluster_profile=data.groupby("Cluster").mean()
cluster_profile
```

Hierarchical clustering is an unsupervised learning algorithm that builds a hierarchy of clusters.

Agglomerative (bottom-up):

1) Starts with each data point as a separate cluster.

2)Iteratively merges the closest clusters based on a distance metric (e.g., Euclidean distance) until all data points belong to a single cluster.

3) This process creates a tree-like structure called a dendrogram, which visually represents the hierarchy of clusters.

Suitablity for iris dataset

  Iris dataset's characteristics—distinct groups, numerical features, and a relatively compact size—align well with the assumptions and workings of the hierarchical clustering algorithm, making it a suitable technique for exploring the structure of this dataset.
